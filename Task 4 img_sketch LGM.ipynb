```python
import cv2
import matplotlib.pyplot as plt

```

```python
# Reading the images

image1 = cv2.imread("flat-white-3402c4f.jpg")
image2 = cv2.imread("images.jpeg")

# cv2.imshow("Coffee", image1)
# cv2.waitKey(0)

#image1
```

```python
cv2.imshow("Tea", image2)
cv2.waitKey(0)
```

```python
# Convert to grayscale

gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# cv2.imshow("Gray_coffee", gray_image1)
# cv2.waitKey(0)
```

```python
# Invert the grayscale image
inverted_image1 = 255 - gray_image1
inverted_image2 = 255 - gray_image2
```

```python

# Apply Gaussian blur to the inverted image
blurred1 = cv2.GaussianBlur(inverted_image1, (21, 21), 0)
blurred2 = cv2.GaussianBlur(inverted_image2, (21, 21), 0)
```

```python

# Invert the blurred image
inverted_blurred1 = 255 - blurred1
inverted_blurred2 = 255 - blurred2
```

```python
# Create pencil sketch by dividing the grayscale image by the inverted blurred image
pencil_sketch1 = cv2.divide(gray_image1, inverted_blurred1, scale=256.0)
pencil_sketch2 = cv2.divide(gray_image2, inverted_blurred2, scale=256.0)
```

```python
# Convert original images to RGB as matploylib uses RGB
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
# Displaying original images and pencil sketches using matplotlib
plt.figure(figsize=(18, 12))
```

```python
plt.subplot(2, 2, 1)
plt.imshow(image1_rgb)
plt.title("Coffee")
plt.axis('off')
```

```python
plt.subplot(2, 2, 2)
plt.imshow(pencil_sketch1, cmap='gray')
plt.title("Coffee Pencil sketch")
plt.axis('off')
```

```python
plt.subplot(2, 2, 3)
plt.imshow(image2_rgb)
plt.title("Tea")
plt.axis('off')
```

```python
plt.subplot(2, 2, 4)
plt.imshow(pencil_sketch2, cmap='gray')
plt.title("Tea sketch")
plt.axis('off')
```
